```python
#List operation
nums = input("user input: ")

numbers = [int(x) for x in nums.split()]

print("\nOriginal List:", numbers)
unique_numbers = []
for n in numbers:
    if n not in unique_numbers:
        unique_numbers.append(n)

print("Removing Duplicates:", unique_numbers)




max_num = unique_numbers[0]
min_num = unique_numbers[0]
total = 0
count = 0

for n in unique_numbers:
    if n > max_num:
        max_num = n
    if n < min_num:
        min_num = n
    total += n
    count += 1

average = total / count

print("\nResults:")
print("Maximum:", max_num)
print("Minimum:", min_num)
print("Average:", average)


```

```python
#Reverse Dictionary
# Step 1️⃣: Original dictionary
fruits = {"apple": 1, "banana": 2, "mango": 3}
print("Original Dictionary:", fruits)

# Step 2️⃣: Create reversed dictionary
reversed_fruits = {}  # খালি dictionary

for key, value in fruits.items():
    reversed_fruits[value] = key  # value becomes key, key becomes value

# Step 3️⃣: Display reversed dictionary
print("Reversed Dictionary:", reversed_fruits)
```

```python
#stack simulation
 #Step 1️⃣: Initialize empty stack
stack = []

# Step 2️⃣: Push 3 elements (take input from user)
for i in range(3):
    elem = input(f"Enter element to push {i+1}: ")
    stack.append(elem)  # push into stack
    print(f"Stack after push {i+1}: {stack}")

# Step 3️⃣: Pop 1 element
popped_element = stack.pop()  # removes last element
print(f"\nPopped Element: {popped_element}")

# Step 4️⃣: Print final stack
print("Final Stack:", stack)
```

```python
#Queue Simulation
# Step 1️⃣: Initialize empty queue
queue = []

# Step 2️⃣: Enqueue 3 elements (user input)
for i in range(3):
    elem = input(f"Enter element to enqueue {i+1}: ")
    queue.append(elem)  # add at the end
    print(f"Queue after enqueue {i+1}: {queue}")

# Step 3️⃣: Dequeue 1 element from front
dequeued_element = queue.pop(0)  # remove element at index 0 (front)
print(f"\nDequeued Element: {dequeued_element}")

# Step 4️⃣: Print final queue
print("Final Queue:", queue)
```

```python
#Function with Default Argument
# Step 1️⃣: Define the function
def simple_interest(p, t, r=5):
    """
    Calculate Simple Interest
    p: principal amount
    t: time in years
    r: rate of interest (default 5%)
    """
    SI = (p * r * t) / 100
    return SI

# Step 2️⃣: Take user input
principal = float(input("Enter Principal amount: "))
time = float(input("Enter Time (in years): "))
rate_input = input("Enter Rate of Interest (press Enter to use default 5%): ")

# Step 3️⃣: Decide whether to use default rate
if rate_input == "":
    SI = simple_interest(principal, time)  # use default r=5
else:
    rate = float(rate_input)
    SI = simple_interest(principal, time, rate)

# Step 4️⃣: Display result
print(f"Simple Interest: {SI}")
```

```python
#Function Returning Multiple Values
# Step 1️⃣: Define the function
def sum_and_difference(a, b):
    """
    Returns sum and difference of two numbers
    """
    sum_val = a + b
    diff_val = a - b
    return sum_val, diff_val  # return multiple values as tuple

# Step 2️⃣: Take user input
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

# Step 3️⃣: Call the function
sum_result, diff_result = sum_and_difference(num1, num2)

# Step 4️⃣: Display the results
print(f"Sum: {sum_result}")
print(f"Difference: {diff_result}")
```

```python
#List Processing with Function
# Step 1️⃣: Define the function
def square_list(numbers):
    """
    Takes a list of numbers and returns a new list with squares
    """
    squared = []
    for num in numbers:
        squared.append(num ** 2)
    return squared

# Step 2️⃣: Take list input from user
numbers = input("Enter numbers separated by space: ").split()
numbers = [float(num) for num in numbers]  # convert each to float

# Step 3️⃣: Call the function
squared_numbers = square_list(numbers)

# Step 4️⃣: Display the result
print("Original List:", numbers)
print("Squared List:", squared_numbers)
```

```python
#oop dstack class 
class Stack:
    def __init__(self):
        self.items = []  # initialize empty stack

    # Push method
    def push(self, item):
        self.items.append(item)
        print(f"Pushed {item}: {self.items}")

    # Pop method
    def pop(self):
        if not self.items:
            print("Stack is empty! Cannot pop.")
            return None
        popped_item = self.items.pop()
        print(f"Popped {popped_item}: {self.items}")
        return popped_item

    # Display method
    def display(self):
        print("Current Stack:", self.items)

# Step 1️⃣: Create Stack object
stack = Stack()

# Step 2️⃣: Push 3 elements
stack.push("A")
stack.push("B")
stack.push("C")

# Step 3️⃣: Pop 1 element
stack.pop()

# Step 4️⃣: Display final stack
stack.display()
```
